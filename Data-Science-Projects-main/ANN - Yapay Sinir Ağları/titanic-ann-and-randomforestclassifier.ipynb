```python
train_df = pd.read_csv('/kaggle/input/titanic/train.csv')
train_df.head(3)
```

```python
train_df.info()
```

```python
train_df = train_df.drop(columns = ['PassengerId' , 'Name' , 'Ticket' , 'Cabin'] , axis = 1)
train_df.head(3)
```

```python
train_df.info()
```

```python
# YaЕџ sГјtunundaki boЕџluklarД± medyan deДџer ile dolduruyoruz
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
train_df.info()
```

```python
# En sД±k biniЕџ yapД±lan limanД± bulup (genelde 'S') boЕџluklara yazД±yoruz
most_freq_port = train_df['Embarked'].mode()[0]
train_df['Embarked'] = train_df['Embarked'].fillna(most_freq_port)
```

```python
train_df.info()
```

```python
train_df = pd.get_dummies(train_df , columns = ['Sex' , 'Embarked'] , drop_first = True , dtype = int)
train_df.head(3)
```

```python
train_df.info()
```

```python
y = train_df['Survived']
x = train_df.drop(columns = ['Survived'] , axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_val , y_train , y_val = train_test_split(x ,y , test_size = 0.2 , random_state = 22)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train_ss = ss.fit_transform(x_train)
x_val_ss = ss.transform(x_val)
```

```python
x_train_ss.shape[1]
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# 1. Model Mimarisi OluЕџturma
model = Sequential([
    # GiriЕџ KatmanД± ve Д°lk Gizli Katman (16 nГ¶ron)
    # input_shape=(8,) Г§ГјnkГј X_train_ss iГ§inde 8 sГјtun var
    Dense(16, activation='relu', input_shape=(x_train_ss.shape[1],)),
    
    # Д°kinci Gizli Katman (8 nГ¶ron)
    Dense(8, activation='relu'),
    
    # Dropout (Ezberlemeyi -Overfitting- Г¶nlemek iГ§in %20'sini kapatД±yoruz)
    Dropout(0.2),
    # Г‡Д±kД±Еџ KatmanД± (Tek nГ¶ron: 0 veya 1 tahmini iГ§in Sigmoid kullanД±lД±r)
    Dense(1, activation='sigmoid')
])

# 2. Modeli Derleme (Compile)
model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

# 3. EДџitim (Fit)
# validation_data kД±smД±na ayД±rdД±ДџД±n val setini veriyoruz
history = model.fit(x_train_ss, y_train, 
                    epochs=100, 
                    batch_size=32, 
                    validation_data=(x_val_ss, y_val),
                    verbose=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
model = rf.fit(x_train_ss , y_train)
model.score(x_val_ss , y_val)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/kaggle/input/heart-disease-prediction-dataset/Heart_Disease_Prediction.csv')
```

```python
df.head(3)
```

```python
df = pd.get_dummies(df , columns = ['Heart Disease'] , drop_first = True , dtype = int)
```

```python
df.head(3)
```

```python
df.info()
```

```python
y = df['Heart Disease_Presence']
x = df.drop(columns = ['Heart Disease_Presence'] , axis = 1)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y ,test_size = 0.3 , random_state = 42)
```

```python
rf = RandomForestClassifier()
model = rf.fit(x_train , y_train)
model.score(x_test,y_test)
```

```python
rf = xgb.XGBClassifier()
model = rf.fit(x_train , y_train)
model.score(x_test , y_test)
```

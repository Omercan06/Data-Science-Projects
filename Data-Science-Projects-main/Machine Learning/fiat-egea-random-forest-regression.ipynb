```python
df = pd.read_csv('/kaggle/input/car-price-turkiye-fiat-egea-2-el-fiyatlar/fiategea.csv')
df.head(3)
```

```python
df = df.drop(columns = ['marka' , 'model' , 'sehir' , 'renk'])
df.head(3)
```

```python
df = pd.get_dummies(df , columns = ['motortipi' , 'paket'] , drop_first = True , dtype = int)
df.head(3)
```

```python
y = df['fiyat']
x = df.drop(columns = ['fiyat'] , axis =1)
x.head(3)
```

```python
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test ,y_train , y_test = train_test_split(x,y,test_size = 0.33 , random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
model.predict([[1.3,2020,100000,0,1,0,0,0,0,0,1]])
```

```python
from sklearn.metrics import mean_absolute_error
real_mae = mean_absolute_error(y_test, y_pred)

print(f"Sade Model Gerçek Hata (TL): {real_mae:,.2f}")
```

```python
df = pd.read_csv('/kaggle/input/insurance/insurance.csv')
df.head(3)
```

```python
df.info()
```

```python
df = pd.get_dummies(df , columns = ['sex' , 'smoker' , 'region'] , drop_first = True , dtype = int)
df.head(3)
```

```python
y = df['charges']
x = df.drop(columns = ['charges'] , axis =1)
```

```python
x.head(3)
```

```python
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y ,test_size = 0.3 , random_state = 6)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(x_train , y_train)
model_pred = model.predict(x_test)
```

```python
df_pred = pd.DataFrame()
df_pred['y_test'] = y_test
```

```python
df_pred['pred'] = model_pred
```

```python
df_pred.head(3)
```

```python
model.score(x_test , y_test)
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error , mean_absolute_percentage_error
```

```python
mean_squared_error(y_test,model_pred)
```

```python
mean_absolute_error(y_test,model_pred)
```

```python
mean_absolute_percentage_error(y_test,model_pred)
```

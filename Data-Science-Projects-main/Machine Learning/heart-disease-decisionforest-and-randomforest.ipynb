```python
from sklearn.tree import DecisionTreeClassifier,export_graphviz
import graphviz
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/kaggle/input/heart-disease-prediction-dataset/Heart_Disease_Prediction.csv')
df.head(3)
```

```python
df = pd.get_dummies(df , columns = ['Heart Disease'] , drop_first = True , dtype = int)
```

```python
df.head(3)
```

```python
df.info()
```

```python
y = df['Heart Disease_Presence']
x = df.drop(columns = ['Heart Disease_Presence'] , axis = 1)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.3 , random_state = 6)
```

```python
tree = DecisionTreeClassifier()
```

```python
model = tree.fit(x_train,y_train)
model.score(x_test , y_test)
```

```python
model.predict([[31,1,2,130,240,0,0,150,0,2,0,0,2]])
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
model = forest.fit(x,y)
model.score(x,y)
```

```python
forest = RandomForestClassifier(n_estimators = 300 , max_depth=4, criterion = 'entropy')
model = forest.fit(x_train , y_train)
model.score(x_test,y_test)
```

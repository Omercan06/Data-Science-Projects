```python
df = pd.read_csv('/kaggle/input/student-depression-and-lifestyle-100k-data/student_lifestyle_100k.csv')
df.head(3)
```

```python
df = df.drop(columns = ['Student_ID'] , axis = 1)
df.info()
```

```python
df = pd.get_dummies(df , columns = ['Gender','Department'], drop_first = True , dtype = int)
df.head(3)
```

```python
df.info()
```

```python
y = df['Depression']
x = df.drop(columns = ['Depression'] , axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, train_size = 0.9 , random_state = 42)
```

```python
rf = RandomForestClassifier()
model = rf.fit(x_train , y_train)
model.score(x_test , y_test)
```

```python
df['Depression'].value_counts()
```

```python
sonuc = model.predict(x_test)
dfs = pd.DataFrame()
dfs['sonuc'] = sonuc
dfs['gercek'] = y_test.values
dfs['sonuc'].value_counts()
```

```python
dfs['gercek'].value_counts()
```

```python
dfs
```

```python
dfs['truebil']=(dfs['gercek']==True) & (dfs['sonuc']==True)
```

```python
dfs['truebil'].value_counts()
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
precision=precision_score(dfs['gercek'],dfs['sonuc'])
```

```python
recall=recall_score(dfs['gercek'],dfs['sonuc'])
```

```python
2*(precision*recall)/(precision+recall)
```

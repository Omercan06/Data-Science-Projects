```python
df = pd.read_csv('/kaggle/input/plane-price-prediction/Plane Price.csv')
df.head(3)
```

```python
df.info()
```

```python
# Virgülleri kaldırıp sayıya çeviren fonksiyon
def clean_and_convert(column):
    # Virgülleri temizle ve sayıya çevir, hata verenleri (string olanları) NaN yap
    return pd.to_numeric(df[column].astype(str).str.replace(',', ''), errors='coerce')

cols_to_fix = ['Max speed Knots', 'HP or lbs thr ea engine', 'All eng rate of climb', 
               'Landing over 50ft', 'Empty weight lbs', 'Range N.M.']

for col in cols_to_fix:
    df[col] = clean_and_convert(col)
```

```python
df.head(3)
```

```python
df.info()
```

```python
df = pd.get_dummies(df , columns = ['Engine Type'], drop_first = True , dtype = int)
```

```python
df = df.drop(columns=['Model Name','Length ft/in' , 'Wing span ft/in'] , axis =1)
```

```python
df.head(3)
df = df.dropna()
```

```python
df.info()
```

```python
y = df['Price']
x = df.drop(columns = ['Price'])
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x ,y ,test_size = 0.3 , random_state = 22)
```

```python
# Y için normalizasyon yapılmaz

# outlier etkisini azaltır

# model performansız artar
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
ss=StandardScaler()
x_train_ss=ss.fit_transform(x_train)
x_test_ss = ss.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
model = rf.fit(x_train , y_train)
model.score(x_test,y_test)
```
